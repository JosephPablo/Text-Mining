```python
import re
# nltk.download('stopwords')
# nltk.download('punkt')
from nltk import word_tokenize 
from nltk.corpus import stopwords
```

```python
REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]')
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
STOPWORDS = set(stopwords.words('english'))
NUMERIC = re.compile("(\w+[0-9]\w+)")
ALPHANUMERIC = re.compile('^[a-zA-Z0-9_]+$')
```

```python
stopwords
```

```python
def text_prepare(doc):

    doc = doc.lower()
    doc = REPLACE_BY_SPACE_RE.sub(' ',doc)
    doc = BAD_SYMBOLS_RE.sub(' ',doc)
    doc = NUMERIC.sub(' ',doc)
    doc = ALPHANUMERIC.sub(' ',doc)
    doc = " ".join([w for w in word_tokenize(doc) if not w in STOPWORDS and len(w) > 2 ])
    return doc
```

```python
new_defect_list = []
for d  in defect_list:
  try:
    new_d = text_prepare(d)
    new_defect_list.append(new_d)
  except AttributeError as e:
#     print(d)
    new_defect_list.append(d)
  
```
